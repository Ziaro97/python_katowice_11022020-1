### Czarno biały filtr

```python
r = im1[:,:,0] + im1[:,:,1] +im1[:,:,2] 
r.shape

```

```python
mean_value =im1.sum(axis=2, keepdims=True) / np.max(mean_value)
plt.imshow(np.ones(3) * mean_value)
```

```python
data = im1.copy()
```

```python
data[:,:,0] = mean_value
data[:,:,1] = mean_value
data[:,:,2] = mean_value
```

```python
plt.imshow(data)
```
